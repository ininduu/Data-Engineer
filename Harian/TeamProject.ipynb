```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import json
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def get_bps_data_selenium(url):
    """Use Selenium to handle JavaScript-rendered content"""
    try:
        print(f"Mengakses URL dengan Selenium: {url}")

        chrome_options = Options()
        chrome_options.add_argument("--headless")
        chrome_options.add_argument("--no-sandbox")
        chrome_options.add_argument("--disable-dev-shm-usage")

        driver = webdriver.Chrome(options=chrome_options)
        driver.get(url)

        try:
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.TAG_NAME, "table"))
            )
        except:
            print("Timeout menunggu tabel dimuat")

        html_content = driver.page_source

        try:
            tables = pd.read_html(html_content)
            if tables and len(tables) > 0:
                df = tables[0]
                print(f"Berhasil mengambil data dengan Selenium: {len(df)} baris")
                driver.quit()
                return df
        except Exception as table_error:
            print(f"Error mencari tabel: {table_error}")

        soup = BeautifulSoup(html_content, 'html.parser')
        scripts = soup.find_all('script')
        table_id = None

        for script in scripts:
            if script.string and 'var tableId' in script.string:
                match = re.search(r'var tableId\s*=\s*[\'"]([^\'"]+)[\'"]', script.string)
                if match:
                    table_id = match.group(1)
                    print(f"Menemukan table ID: {table_id}")
                    break

        data_params = {}
        for script in scripts:
            if script.string and 'var data' in script.string:
                try:
                    json_match = re.search(r'var data\s*=\s*(\{.*?\});', script.string, re.DOTALL)
                    if json_match:
                        data_json = json_match.group(1)
                        data_json = re.sub(r'(\w+):', r'"\1":', data_json)
                        data_json = data_json.replace("'", '"')
                        try:
                            data_params = json.loads(data_json)
                            print(f"Found data parameters: {data_params.keys()}")
                        except json.JSONDecodeError:
                            print("Failed to parse data parameters")
                except Exception as e:
                    print(f"Error parsing script: {e}")

        driver.quit()

        if table_id:
            return try_api_calls(table_id, data_params)

        print("Tidak dapat menemukan tabel dengan Selenium")
        return pd.DataFrame()

    except Exception as e:
        print(f"Error saat menggunakan Selenium: {e}")
        return pd.DataFrame()

def try_api_calls(table_id, data_params=None):
    """Try various API endpoints to get table data"""
    api_urls = [
        f"https://webapi.bps.go.id/v1/api/view/domain/3500/model/statictable/lang/ind/id/2521/key/{API_KEY}",
    ]

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        'Accept': 'application/json, text/javascript, */*; q=0.01',
        'Accept-Language': 'en-US,en;q=0.9',
        'Referer': 'https://jatim.bps.go.id/id/statistics-table/1/MjUyMSMx/luas-panen-produktivitas-dan-produksi-padi-menurut-kabupaten-kota-di-provinsi-jawa-timur-2021-dan-2022.html',
        'X-Requested-With': 'XMLHttpRequest'
    }

    for api_url in api_urls:
        print(f"Mencoba API: {api_url}")

        try:
            time.sleep(1)

            api_response = requests.get(api_url, headers=headers)

            if api_response.status_code == 200:
                try:
                    data = api_response.json()

                    html_table = data.get('content', '')

                    if html_table:
                        try:
                            dfs = pd.read_html(html_table)
                            if len(dfs) > 0:
                                df = dfs[0]
                                print(f"Berhasil mengambil data dengan API: {len(df)} baris")
                                return df
                        except Exception as e:
                            print(f"Error konversi HTML ke DataFrame: {e}")
                except Exception as e:
                    print(f"API mengembalikan response non-JSON atau error: {e}")
                    if "<table" in api_response.text:
                        try:
                            dfs = pd.read_html(api_response.text)
                            if len(dfs) > 0:
                                df = dfs[0]
                                print(f"Berhasil mengambil data mentah dari API: {len(df)} baris")
                                return df
                        except Exception as table_err:
                            print(f"Error parsing tabel dari response mentah: {table_err}")
        except Exception as request_error:
            print(f"Error request API: {request_error}")

    print("Semua metode API gagal")
    return pd.DataFrame()

def get_bps_data_alternative(url):
    """Try alternative approach by directly accessing embed URL"""
    try:
        print(f"Mencoba metode alternatif untuk URL: {url}")

        url_match = re.search(r'/([A-Za-z0-9=]+)/[^/]+$', url)
        if not url_match:
            print("Tidak dapat menemukan ID dari URL")
            return pd.DataFrame()

        table_id = url_match.group(1)

        embed_url = f"https://jatim.bps.go.id/id/statistics-table/1/MjUyMSMx/luas-panen-produktivitas-dan-produksi-padi-menurut-kabupaten-kota-di-provinsi-jawa-timur-2021-dan-2022.html"
        print(f"Mencoba URL embed: {embed_url}")

        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        }

        response = requests.get(embed_url, headers=headers)

        if response.status_code == 200:
            try:
                dfs = pd.read_html(response.content)
                if len(dfs) > 0:
                    df = dfs[0]
                    print(f"Berhasil mengambil data dari embed URL: {len(df)} baris")
                    return df
            except Exception as e:
                print(f"Error parsing embed URL: {e}")

        print("Metode alternatif gagal")
        return pd.DataFrame()

    except Exception as e:
        print(f"Error pada metode alternatif: {e}")
        return pd.DataFrame()

def get_bps_data(url):
    """Main function to try multiple methods to get data"""
    df = get_bps_data_selenium(url)
    if not df.empty:
        return df

    df = get_bps_data_alternative(url)
    if not df.empty:
        return df

    print("Semua metode gagal untuk URL ini")
    return pd.DataFrame()

def save_to_csv(df, output_file):
    try:
        df.to_csv(output_file, index=False, sep=",", encoding="utf-8")
        print(f"Data berhasil disimpan ke {output_file}")
        return True
    except Exception as e:
        print(f"Error menyimpan ke CSV {output_file}: {e}")
        return False

def display_csv(csv_file):
    try:
        df = pd.read_csv(csv_file)
        print(f"\nIsi dari {csv_file}:")
        print(df.head())
        return df
    except Exception as e:
        print(f"Error menampilkan CSV {csv_file}: {e}")
        return pd.DataFrame()

def process_all_urls():
    urls = [
        ("https://jatim.bps.go.id/id/statistics-table/1/MjUyMSMx/luas-panen-produktivitas-dan-produksi-padi-menurut-kabupaten-kota-di-provinsi-jawa-timur-2021-dan-2022.html", "data_JatimPadi.csv")
    ]

    results = []

    for url, output_file in urls:
        print(f"\n=== Memproses {url} ===")
        df = get_bps_data(url)

        if not df.empty:
            save_to_csv(df, output_file)
            display_csv(output_file)
            results.append((output_file, df))
        else:
            print(f"Gagal mengambil data dari {url}")

    return results

process_all_urls()
```
